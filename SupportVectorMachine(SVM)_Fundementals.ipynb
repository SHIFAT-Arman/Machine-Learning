```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

# Creating the data

```python
X, y = make_blobs(n_samples = 100, n_features = 2, centers = 2, cluster_std = 1.5, random_state = 42)
```

## Visualize the data

```python
# plotting the data
plt.scatter(X[:,0], X[:,1], c = y, cmap = 'coolwarm')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(linestyle='--', alpha=0.3)
plt.show()
```

# Draw the best fit line (y = mx + b) to separate the class


```python
# plotting the data
plt.scatter(X[:,0], X[:,1], c = y, cmap = 'coolwarm')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(linestyle='--', alpha=0.3)

x = np.linspace(-6,6)
list = [(1,3),(1,4),(1,5),(1,6), (1,7)]
for m, b in list:
    y_ = m * x + b
    plt.plot(x, y_, 'm', lw =3)

```

# Importing SVM

```python
from sklearn.svm import SVC
svc = SVC()
```

- C = Regularization parameter. The strength of the regularization is inversely proportional to C.
This C is same as K we studied in the theory of SVM. Larger the value of K, Larger will be the margin and the support vectors are farther away from the hyperplane and the margin is hard.

## Large value for K

```python
svc = SVC(kernel = 'linear', C = 1)
svc.fit(X, y)
```

# Drawing Hyperplane of SVM

```python
from sklearn.inspection import DecisionBoundaryDisplay

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    svc,
    X,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    svc.support_vectors_[:, 0],
    svc.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

## Small value for K

```python
svc = SVC(kernel = 'linear', C = 0.01)
svc.fit(X, y)
```

```python

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    svc,
    X,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    svc.support_vectors_[:, 0],
    svc.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

# Kernel trick
- Kernel SVM ( kernel = ' rbf ', gamma = ' scale ' ). We will explore gamma function when the kernel is "rbf". gamma defines how much influence each training sample has.

```python
k_svc = SVC(kernel = 'rbf', gamma = 'scale')
k_svc.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    k_svc,
    X,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    k_svc.support_vectors_[:, 0],
    k_svc.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

Kernel SVM ( kernel = ' rbf ', gamma = ' auto ' )

```python
k_svc = SVC(kernel = 'rbf',C=1, gamma = 'auto') # When gamma = auto. It tries to learn the noise of the data
                                                   # and enter into the overfitting.
k_svc.fit(X, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    k_svc,
    X,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    k_svc.support_vectors_[:, 0],
    k_svc.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

# Data is not linearly seperable

```python
from sklearn.datasets import make_circles
Xc, yc = make_circles(n_samples=100, noise=0.05,factor=0.1, random_state=42)
```

## Visualize the data

```python
# plotting the data
plt.scatter(Xc[:,0], Xc[:,1], c = yc, cmap = 'coolwarm')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(linestyle='--', alpha=0.3)
plt.show()
```

# Applying SVM

```python
c_svm = SVC(kernel = 'rbf', gamma='auto')
c_svm.fit(Xc, yc)
```

```python
plt.scatter(Xc[:, 0], Xc[:, 1], c=yc, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    c_svm,
    Xc,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    c_svm.support_vectors_[:, 0],
    c_svm.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```
