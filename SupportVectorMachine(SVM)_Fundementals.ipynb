```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

# Creating the data

```python
X, y = make_blobs(n_samples = 100, n_features = 2, centers = 2, cluster_std = 1.5, random_state = 42)
```

# Visualize the data

```python
# plotting the data
plt.scatter(X[:,0], X[:,1], c = y, cmap = 'coolwarm')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(linestyle='--', alpha=0.3)
plt.show()
```

# Draw the best fit line (y = mx + b) to separate the class


```python
# plotting the data
plt.scatter(X[:,0], X[:,1], c = y, cmap = 'coolwarm')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(linestyle='--', alpha=0.3)

x = np.linspace(-6,6)
list = [(1,3),(1,4),(1,5),(1,6), (1,7)]
for m, b in list:
    y_ = m * x + b
    plt.plot(x, y_, 'm', lw =3)

```

# Importing SVM

```python
from sklearn.svm import SVC
svc = SVC()
```

- C = Regularization parameter. The strength of the regularization is inversely proportional to C.
This C is same as K we studied in the theory of SVM. Larger the value of K, Larger will be the margin and the support vectors are farther away from the hyperplane and the margin is hard.

## Large value for K

```python
svc = SVC(kernel = 'linear', C = 1)
svc.fit(X, y)
```

# Drawing Hyperplane of SVM

```python
from sklearn.inspection import DecisionBoundaryDisplay

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    svc,
    X,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    svc.support_vectors_[:, 0],
    svc.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

## Small value for K

```python
svc = SVC(kernel = 'linear', C = 0.01)
svc.fit(X, y)
```

```python

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='coolwarm')

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    svc,
    X,
    plot_method="contour",
    colors="r",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    svc.support_vectors_[:, 0],
    svc.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```
